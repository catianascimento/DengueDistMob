## Imports

```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
method_name = "pearson"
# method_name = "spearman"

cases_filename = "cases_matrix_all.xlsx"
# cases_filename = "cases_matrix_2014-2015.xlsx"
# cases_filename = "daily_cases_with_zeros_no_dates.xlsx"
# cases_filename = "daily_cases_with_zeros_no_dates_1415.xlsx"
data = pd.read_excel(cases_filename) 
data
```

```python
#Removing "Vazios urbanos"
data = data.drop([36, 41, 50, 52, 55], axis=1)
data.head()
```

##Calculating Correlation


```python
dt = data.astype(float)

correlations = dt.corr(method=method_name)

plt.figure(figsize = (20,20))

ax = sns.heatmap(
    correlations, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=400),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.title('Correlation Coefficient Matrix', fontsize=16);
plt.savefig(method_name +'_' + 'correlation.pdf')

# export_csv = correlations.to_csv('correlations.csv', index = True, header=True)
```

```python
correlations
```

Creating dataframe of correlations

```python
index = correlations.index

x1x2correlations = pd.DataFrame(columns=['X1','X2','corr'])

for i in index:
  for j in index:
    if(j>i):
      new_row = {'X1':i, 'X2':j, 'corr':correlations._get_value(i, j)}
      x1x2correlations = x1x2correlations.append(new_row, ignore_index=True)
      
x1x2correlations['X1'] = x1x2correlations['X1'].astype('int')
x1x2correlations['X2'] = x1x2correlations['X2'].astype('int')

x1x2correlations.to_csv(method_name+  '_' +'x1x2_correlations.csv', index=False)

x1x2correlations
```

Cross-Correlation

```python
#The delay is always applied over the second timeseries
def crosscorr(datax, datay, lag=0, method_name='pearson'):
  if (lag!=0):
      shiftedy = datay.shift(lag)
      shiftedy.iloc[:lag] = datay.iloc[-lag:].values
      corr = datax.corr(shiftedy, method = method_name)
      return corr
  else: 
      return datax.corr(datay.shift(lag), method = method_name)
```

```python
ks = list(range(-8,9))
corrs = []
for k in ks:
  corrs.append(crosscorr(data.iloc[:,6], data.iloc[:,12], k, method_name))
```

```python
df= pd.DataFrame()

df['ks'] = ks
df['corrs'] = corrs
```

Calculating Max Correlation for all of pair of TZs

```python
ks = list(range(-8,9))

indexes = correlations.columns

max_corrs = pd.DataFrame().reindex_like(correlations)
max_ks = pd.DataFrame().reindex_like(correlations)

# max_corrs
for i in indexes:
  for j in indexes: #we don't need to fill all matrix
    # if(j>=i):
      max_k=0
      max_corr=-2
      ks = list(range(-8,9))
      for k in ks:
        corr = crosscorr(data[i], data[j], k, method_name)
        if(max_corr < corr): #Nao considerar o valor em modulo
           max_corr = corr
           max_k = k
      max_corrs.loc[i,j]=max_corr
      max_ks.loc[i,j]=max_k
```
