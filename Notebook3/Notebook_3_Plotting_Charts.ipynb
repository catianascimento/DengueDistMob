##Imports

```python
import pandas as pd
import numpy as np
```

```python
#Initial parameters
limit_dist = 3000
limit_mob = 2000

mob_mtx_filename = 'mob_matrix.csv'
# dist_mtx_filename = 'distance_matrix.csv'
dist_mtx_filename = 'distance_matrix_norm.csv'
corrs_filename = 'pearson_x1x2_correlations.csv'
max_corrs_filename = 'pearson_x1x2_max_corrs.csv'
# corrs_filename = 'spearman_x1x2_correlations.csv'
# max_corrs_filename = 'spearman_x1x2_max_corrs.csv'
```

```python
#Reading files 

distance_mtx = dist_mtx_filename
mob_mtx = mob_mtx_filename

corrs = corrs_filename
max_corrs = max_corrs_filename
```

```python
#Eliminating unnecessary rows and columns from matrix
dist = pd.read_csv(distance_mtx, delimiter=';', index_col = 'ZT')

# dist = dist.div(1000, axis=0) #converting km - m
dist.columns = dist.index
dist
```

```python
correlations = pd.read_csv(corrs)
correlations = correlations.astype({"X1": int, "X2": int})
correlations
```

```python
mob = pd.read_csv(mob_mtx, delimiter=';', index_col = 'x')
mob.columns = mob.index
mob.head()
```

```python
max_corrs_df = pd.read_csv(max_corrs)

max_corrs_df = max_corrs_df.astype({"X1": int, "X2": int})
max_corrs_df.head()
```

```python
correlations['dist'] = correlations.apply(lambda x: dist.loc[int(x[0]), int(x[1])], axis=1)

dist_versus_correlations = correlations[['dist', 'corr']]
dist_versus_correlations = dist_versus_correlations.dropna()
dist_versus_correlations = dist_versus_correlations[dist_versus_correlations.dist != 0.000]
dist_versus_correlations.head()
```

```python
"""###Sorting daframe"""

dist_versus_correlations = dist_versus_correlations.sort_values(by ='dist')
```

```python
import math

limit = limit_dist

max_dist = dist_versus_correlations.loc[dist_versus_correlations['dist'].idxmax()]
ceil = math.ceil(max_dist['dist'])+limit
```

```python
max_corrs = []
min_corrs = []
mean_corrs = []

for i in range(0, ceil, limit):
  interval = dist_versus_correlations[dist_versus_correlations['dist'].between(i, i+limit)]
  
  if(len(interval)>0):
    corr_mean = interval['corr'].mean()
    dist_mean = interval['dist'].mean()
    num_elements = len(interval)
    mean_corrs.append((dist_mean,corr_mean, num_elements))
    
    max = interval.loc[interval['corr'].idxmax()] #max contains a pair (dist_max_corr, max_corr)
    min = interval.loc[interval['corr'].idxmin()]
    max_corrs.append((dist_mean, max['corr']))
    min_corrs.append((dist_mean, min['corr']))
```

```python
mins_dataframe = pd.DataFrame(min_corrs)
maxs_dataframe = pd.DataFrame(max_corrs)
means_dataframe = pd.DataFrame(mean_corrs)

means_dataframe.rename(columns={0: 'dist', 1: 'corr', 2: 'num_elements'}, inplace=True)
maxs_dataframe.rename(columns={0: 'dist', 1: 'corr'}, inplace=True)
mins_dataframe.rename(columns={0: 'dist', 1: 'corr'}, inplace=True)
```

Distance X Correlation Coefficient - using plotly

```python
dist_versus_correlations = dist_versus_correlations.groupby(dist_versus_correlations.columns.tolist()).size().reset_index().\
    rename(columns={0:'count'})
dist_versus_correlations.head()
```

```python
num_splits=40
dfs = np.array_split(dist_versus_correlations, num_splits)
mins= []
maxs= []
means= []
medians = []
for i in range(num_splits):
  mins.append(dfs[i]['corr'].min())
  maxs.append(dfs[i]['corr'].max())
  means.append(dfs[i]['corr'].mean())
  medians.append(dfs[i]['dist'].median())

d1 = {'mins':mins,'maxs':maxs,'means':means,'medians':medians}
data1 = pd.DataFrame(d1)

data1.head()
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=dist_versus_correlations['dist'], y=dist_versus_correlations['corr'],
                    mode='markers', name='Original Data', 
                     hovertemplate =
                    '<i>Distance</i>: %{x}<br>'+
                    '<i>Max Corr</i>: %{y}<br>'+
                    '<i>Count</i>: %{marker.size:,}',
                    marker=dict(
                       opacity=0.25,
                       color=dist_versus_correlations['corr'],
                       colorbar=dict(
                          title="corrs"
                       ),       
                       size=dist_versus_correlations['count'],
                       sizemode='area',
                       sizeref=2.*(dist_versus_correlations['count'].max())/(20.**2),
                       sizemin=1,             
                       colorscale="Bluered"
                    )))
fig2.update_layout(legend= {'itemsizing': 'constant'})

fig2.add_trace(go.Scatter(x=data1['medians'], y=data1['maxs'],
                    mode='lines+markers', name='max', line=dict(color='rgb(130,8,8)', width=2)))
fig2.add_trace(go.Scatter(x=data1['medians'], y=data1['means'],
                    mode='lines+markers', name='mean', line=dict(color='rgb(100,50,100)', width=2)))
fig2.add_trace(go.Scatter(x=data1['medians'], y=data1['mins'],
                    mode='lines+markers', name='min', line=dict(color='rgb(8,8,130)', width=2)))


fig2.update_xaxes(type="log")
fig2.update_layout(
    title_text="Distance x Correlation Coefficient",
    xaxis_title="Distance(km)",
    yaxis_title="Correlation Coefficient",
    font=dict(
        family="Arial",
        size=15,
        color="RebeccaPurple"
    ),
    width=1000,
    height=600,
    legend=dict(
      yanchor="top",
      y=0.99,
      xanchor="left",
      x=0.05
    ),
    hovermode="closest"
   )

fig2.show()
```

# Linear Regression

Therefore, a linear regression was performed to assess the possible linearity relationship between correlation x distance" and correlation x mobility". This method consists of estimating parameters b0 and b 1 of the following linear equation (X): Y = b1x+b0.

For the statistical treatment of data on distance, mobility, and correlation of Dengue cases presented in this work, a hypothesis test was developed for the parameters calculated from the linear regression line of the data. The most conservative hypothesis, H0, states that there is no statistically signi cant linear relationship between the 2 variables studied and which is sought to be rejected. Therefore, a useful approach is to use the p-value or probability of signi cance. The p-value is a statistic commonly used to synthesize the result of a hypothesis test and indicates the probability of the test statistic being equal to or more extreme than the sample statistic, that is, of being in the rejection region of H0, assuming it is true. If p < alpha , then a type I error occurred and H0 is rejected in favor of the alternative hypothesis H1, more innovative.

```python
from scipy.stats import t 
import numpy as np
import scipy


def teste_hipotese(x,y,alfa=0.05):
    
    
    '''T-test hypothesis testing for regression. Returns 1 if null Hypthotesis can be rejected (considering the chosen alfa )
    and 0 if it can't.
    
    Parameters:
    -------------------------------------------
    x and y: Variables 
    alfa: Desired statistical significance.'''
    
    n = len(x)
    xmed = np.sum(x)/n;
    ymed = np.sum(y)/n;
    i = 0
    ttable = 1.960
    reject=0       
    reject0=0     
    Sxy = 0
    Sxx = 0 
    Syy = 0
    for i in range(n):
        Sxy = Sxy + ((x[i]-xmed)*(y[i]-ymed))
        Sxx = Sxx + np.power(x[i]-xmed, 2)

    #b1hat and b0hat
    b1 = Sxy/Sxx
    b0 = ymed-b1*xmed

    # print(b1)
    # print(b0)

    for i in range(n):
        Syy = Syy + np.power(y[i]-ymed, 2)

    R2 = np.power(Sxy,2)/(Sxx*Syy)
    R2a = 1 - ((n-1)/(n-2))*(1-R2)
    QME = (Syy-(b1*Sxy))/(n-2)
    bla = (1.0/n) + (xmed*xmed)/(Sxx)
    t0 = b0 / np.sqrt(QME*bla)
    t = b1 / np.sqrt(QME/Sxx)
    
        
    #T-Student
    # pvalue = 2*(1 - scipy.stats.t.cdf(np.abs(t),df=n-2))

    # Z - Normal
    pvalue = 2*(1 - scipy.stats.norm.cdf(np.abs(t),))

    if np.abs(pvalue)<=alfa:
        reject=1
    else:
        reject=0
    print("b0: "+str(b0))
    print("b1: "+str(b1))
    # print(np.sqrt(QME/Sxx))
    print("t: "+str(t))
    # print(Sxy)
    print("reject: "+str(reject))
    print("pvalue: "+str(pvalue))
    return reject,pvalue
```
